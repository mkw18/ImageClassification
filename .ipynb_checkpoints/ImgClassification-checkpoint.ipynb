```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import transforms, utils
from torch.autograd import Variable
import torch.utils.data as Data
import torchvision.models as models
from model.resnet import ResNet18, ResNet34, ResNet50, ResNet101, ResNet152, ResNext50, ResNext101, wide_ResNet50, wide_ResNet101
from torchvision.models import resnet18, resnet34, resnet50, resnet101, resnet152, resnext50_32x4d, resnext101_32x8d, wide_resnet50_2, wide_resnet101_2
import numpy as np
import pandas
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.metrics import accuracy_score, f1_score, roc_curve, precision_recall_curve, average_precision_score, auc
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, matthews_corrcoef, roc_auc_score
import copy
from sklearn.model_selection import train_test_split

#重写数据库函数
class TrainDataset(Data.Dataset):
    def __init__(self, image, label, transform):
        self.Image = image
        self.Label = label
        self.size = len(label)
        self.transform = transform

    def __len__(self):
        return self.size

    def __getitem__(self, idx):
        image = self.Image[idx]
        label = self.Label[idx]
        if self.transform:
            image = self.transform(image)
        return image, label


class TestDataset(Data.Dataset):
    def __init__(self, image, transform):
        self.data = image
        self.size = len(image)
        self.transform = transform

    def __len__(self):
        return self.size

    def __getitem__(self, idx):
        sample = self.data[idx]
        if self.transform:
            sample = self.transform(sample)
        return (sample)
    
class LabelSmoothingCrossEntropy(nn.Module):
     def __init__(self, eps=0.1, reduction='mean'):
          super(LabelSmoothingCrossEntropy, self).__init__()
          self.eps = eps
          self.reduction = reduction

     def forward(self, output, target):
         c = output.size()[-1]
         log_preds = F.log_softmax(output, dim=-1)
         if self.reduction=='sum':
              loss = -log_preds.sum()
         else:
             loss = -log_preds.sum(dim=-1)
             if self.reduction=='mean':
                  loss=loss.mean()
         return loss * self.eps/c + (1-self.eps) * F.nll_loss(log_preds, target, reduction=self.reduction)
```

```python
#构建训练集和验证集
BATCH_SIZE = 128
train_np = np.load("./train.npy")
train_result = pandas.read_csv("./train1.csv", header=0)
train_result = np.asarray(train_result.iloc[:, 1])
train_result2 = pandas.read_csv("./train2.csv", header=0)
train_result2 = np.asarray(train_result2.iloc[:, 1])
train_coarse,test_coarse,train_coarse_label,test_coarse_label = train_test_split(train_np,train_result,test_size=0.2,random_state=5)
validate_coarse,test_coarse,validate_coarse_label,test_coarse_label = train_test_split(test_coarse,test_coarse_label,test_size=0.2,random_state=0)
train_fine,test_fine,train_fine_label,test_fine_label = train_test_split(train_np,train_result2,test_size=0.2,random_state=5)
validate_fine,test_fine,validate_fine_label,test_fine_label = train_test_split(test_fine,test_fine_label,test_size=0.2,random_state=0)
train_coarse = np.reshape(train_coarse, (train_coarse.shape[0], 3, 32, 32))
validate_coarse = np.reshape(validate_coarse, (validate_coarse.shape[0], 3, 32, 32))
test_coarse = np.reshape(test_coarse,(test_coarse.shape[0], 3, 32, 32))
train_fine = np.reshape(train_fine, (train_fine.shape[0], 3, 32, 32))
validate_fine = np.reshape(validate_fine,(validate_fine.shape[0], 3, 32, 32))
test_fine = np.reshape(test_fine,(test_fine.shape[0], 3, 32, 32))

train_coarse_t = torch.tensor(train_coarse)
validate_coarse_t = torch.tensor(validate_coarse)
test_coarse_t = torch.tensor(test_coarse)
train_fine_t = torch.tensor(train_fine)
validate_fine_t = torch.tensor(validate_fine)
test_fine_t = torch.tensor(test_fine)
train_coarse_label_t = torch.tensor(train_coarse_label)
validate_coarse_label_t = torch.tensor(validate_coarse_label)
test_coarse_label_t = torch.tensor(test_coarse_label)
train_fine_label_t = torch.tensor(train_fine_label)
validate_fine_label_t = torch.tensor(validate_fine_label)
test_fine_label_t = torch.tensor(test_fine_label)
mean = np.array([0.49139968, 0.48215827, 0.44653124])
std = np.array([0.2023, 0.1994, 0.2010])
transform = transforms.Compose([
    transforms.ToPILImage(), 
    transforms.Resize(224),
    transforms.RandomHorizontalFlip(), 
    transforms.ToTensor(),
    transforms.Normalize(mean, std),
])
transform1 = transforms.Compose([
    transforms.ToPILImage(), 
    transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])
train_coarse_dataset = TrainDataset(train_coarse_t, train_coarse_label_t, transform)
train_fine_dataset = TrainDataset(train_fine_t, train_fine_label_t, transform)
validate_coarse_dataset = TrainDataset(validate_coarse_t, validate_coarse_label_t, transform1)
validate_fine_dataset = TrainDataset(validate_fine_t, validate_fine_label_t, transform1)
test_coarse_dataset = TrainDataset(test_coarse_t, test_coarse_label_t, transform1)
test_fine_dataset = TrainDataset(test_fine_t, test_fine_label_t, transform1)
```

```python
# 引出网络，20分类
num_class = 20
net = resnet101(pretrained = True)
fc_features = net.fc.in_features
net.fc = nn.Linear(fc_features, num_class)

init_lr = 0.01
optimizer = optim.SGD(net.parameters(), lr=init_lr, momentum=0.9, weight_decay=1e-4)
#optimizer = optim.Adam(net.parameters(), lr=init_lr, weight_decay=1e-4)
losses = []
acces = []
validate_loss = []
validate_acc = []
# 以下为针对不同resnet可以有的交叉熵损失，通过前一轮训练得出的
w18 = torch.tensor([1.09006116, 0.94298942, 0.94549072, 0.93802632, 0.93311518,
       0.93802632, 0.9739071 , 0.90931122, 1.07042042, 0.98466851,
       0.98739612, 0.96598916, 1.04838235, 1.09340491, 0.99567039,
       1.10356037, 1.10698758, 0.97125341, 0.99567039, 1.0867378 ])
w34 = torch.Tensor([1.05338235, 0.91132316, 0.96536388, 0.93756545, 0.9425    ,
       0.94498681, 0.97588556, 0.92069409, 1.06592262, 0.96797297,
       1.01458924, 0.9525266 , 1.02916667, 1.09525994, 1.00887324,
       1.06910448, 1.102     , 0.98663912, 1.03511561, 1.09192073])
wx50 = torch.tensor([1.06205882, 0.95529101, 0.96037234, 0.92827763, 0.94528796,
       0.92589744, 0.98392371, 0.9306701 , 1.0683432 , 0.97069892,
       1.03467049, 0.95782493, 1.03764368, 1.04364162, 1.00305556,
       1.08438438, 1.08113772, 0.99751381, 1.02585227, 1.05894428])
# 以下两种交叉熵二选一
#loss = nn.CrossEntropyLoss()
loss = LabelSmoothingCrossEntropy()
net = net.to(device)
loss = loss.to(device)
```

```python
#开始训练
init = 0
print("resnet101 Start Training...")
maxacc = 0.0
maxepoch = 0
f = open('train_record/resnet101_init_lr%f_SGD_MLR_60epoch_weight.txt' % init_lr, 'w')
f1 = open('validate_record/resnet101_init_lr%f_SGD_MLR_60epoch_weight.txt' % init_lr, 'w')
f.write("resnet101\n")
f1.write("resnet101\n")
f1.write("epoch\t")
for i in range(num_class):
    f1.write("%d\t"%(i+1))
f1.write("\n")
scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones = [15,25,35,45], gamma = 0.2)
epochs = 60
acc_all = np.zeros((epochs, num_class))
for epoch in range(init, epochs):
    train_loss = 0.0
    train_acc = 0.0
    net.train()

    for i, data in enumerate(train_coarse_loader):
        inputs, labels = data
        inputs = inputs.float()
        inputs = Variable(inputs)
        labels = Variable(labels)
        inputs = inputs.to(device)
        labels = labels.to(device)
        outputs = net(inputs)
        lossess = loss(outputs, labels)
        optimizer.zero_grad()
        lossess.backward()
        optimizer.step()

        train_loss += float(lossess)

        _, pred = outputs.max(1)
        num_correct = (pred == labels).sum()
        acc = num_correct
        train_acc += acc

    losses.append(train_loss / len(train_coarse_loader))
    acces.append(train_acc / train_coarse.shape[0])

    #求得验证集的正确率
    cross_acc = 0.0
    cross_loss = 0.0
    net.eval()
    with torch.no_grad():
        for data in validate_coarse_loader:
            inputs, labels = data
            inputs = inputs.float()
            inputs = Variable(inputs)
            labels = Variable(labels)
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = net(inputs)
            lossess = loss(outputs, labels)
            cross_loss += float(lossess)
            _, pred = outputs.max(1)
            num_correct = (pred == labels).sum()
            acc = num_correct
            cross_acc += acc
            for i in range(labels.shape[0]):
                if pred[i] == labels[i]:
                    acc_all[epoch,labels[i]] += 1
    validate_loss.append(cross_loss / len(validate_coarse_loader))
    validate_acc.append(cross_acc / validate_coarse.shape[0])
    if validate_acc[-1] > maxacc:
        maxacc = validate_acc[-1]
        maxepoch = epoch
        torch.save({'epoch': epoch,
                    'model_state_dict': net.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'loss': loss}, 'weights/weights_resnet101_init_lr%f_SGD_MLR_60epoch_weight.tar' % init_lr)
    print("epoch: %d train_loss: %f train_acc: %f validate_loss: %f validate_acc: %f lr: %f" % (epoch+1, losses[-1], acces[-1], validate_loss[-1], validate_acc[-1], scheduler.get_last_lr()[0]))
    f.write("epoch: %d train_loss: %f train_acc: %f validate_loss: %f validate_acc: %f lr: %f\n" % (epoch+1, losses[-1], acces[-1], validate_loss[-1], validate_acc[-1], scheduler.get_last_lr()[0]))
    acc_all[epoch,:] /= validate_coarse.shape[0] / num_class
    f1.write("%d\t" % epoch)
    for i in range(num_class):
        f1.write("%f\t" % acc_all[epoch, i])
    f1.write("\n")
    if ((epoch+1) % 10 == 0):
        f.write("Total epoch is %d, the max accuracy is %f, when epoch is %d.\n" % (epoch+1, maxacc, maxepoch+1))
    f.flush()
    f1.flush()
    scheduler.step()
print("Done Training!")
f.write("Done!\n")
f.write("The max accuracy is %f, when epoch is %d.\n" % (maxacc, maxepoch+1))
f.close()
acc_mean = np.mean(acc_all[20:,:],axis=1)
f1.write("mean:\t")
for i in range(num_class):
    f1.write("%f\t"%acc_mean[i])
f1.write("\n")
f1.close()
```

```python
# 引出网络，100分类
num_class = 100
net = resnet50(pretrained = True)
fc_features = net.fc.in_features
net.fc = nn.Linear(fc_features, num_class)

init_lr = 0.01
optimizer = optim.SGD(net.parameters(), lr=init_lr, momentum=0.9, weight_decay=1e-4)
losses = []
acces = []
validate_loss = []
validate_acc = []
loss = nn.CrossEntropyLoss()
#loss = LabelSmoothingCrossEntropy()
net = net.to(device)
loss = loss.to(device)
```

```python
#开始训练
init = 0
print("resnet50 Start Training...")
maxacc = 0.0
maxepoch = 0
f = open('train_record/resnet50_init_lr%f_SGD_MLR_50epoch100mp.txt' % init_lr, 'w')
f1 = open('validate_record/resnet50_init_lr%f_SGD_MLR_50epoch100mp.txt' % init_lr, 'w')
f.write("resnet50\n")
f1.write("resnet50\n")
f1.write("epoch\t")
for i in range(num_class):
    f1.write("%d\t"%(i+1))
f1.write("\n")
scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones = [15,25,35,45], gamma = 0.2)
epochs = 50
acc_all = np.zeros((epochs, num_class))
for epoch in range(init, epochs):
    train_loss = 0.0
    train_acc = 0.0
    net.train()

    for i, data in enumerate(train_fine_loader):
        inputs, labels = data
        inputs = inputs.float()
        inputs = Variable(inputs)
        labels = Variable(labels)
        inputs = inputs.to(device)
        labels = labels.to(device)
        
        optimizer.zero_grad()

        alpha = 0.2
        lam = np.random.beta(alpha,alpha)
        index = torch.randperm(inputs.size(0)).to(device)
        inputs_2 = lam*inputs + (1-lam)*inputs[index,:]
        targets_a, targets_b = labels, labels[index]
        outputs = net(inputs_2)
        lossess = lam*loss(outputs, targets_a) + (1-lam) * loss(outputs, targets_b)
        
        lossess.backward()
        optimizer.step()

        train_loss += float(lossess)

        _, pred = outputs.max(1)
        num_correct = (pred == labels).sum()
        acc = num_correct
        train_acc += acc

    losses.append(train_loss / len(train_fine_loader))
    acces.append(train_acc / train_fine.shape[0])

    #求得验证集的正确率
    cross_acc = 0.0
    cross_loss = 0.0
    net.eval()
    with torch.no_grad():
        for data in validate_fine_loader:
            inputs, labels = data
            inputs = inputs.float()
            inputs = Variable(inputs)
            labels = Variable(labels)
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = net(inputs)
            lossess = loss(outputs, labels)
            cross_loss += float(lossess)
            _, pred = outputs.max(1)
            num_correct = (pred == labels).sum()
            acc = num_correct
            cross_acc += acc
            for i in range(labels.shape[0]):
                if pred[i] == labels[i]:
                    acc_all[epoch,labels[i]] += 1
    validate_loss.append(cross_loss / len(validate_fine_loader))
    validate_acc.append(cross_acc / validate_fine.shape[0])
    if validate_acc[-1] > maxacc:
        maxacc = validate_acc[-1]
        maxepoch = epoch
        torch.save({'epoch': epoch,
                    'model_state_dict': net.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'loss': loss}, 'weights/weights_resnet50_init_lr%f_SGD_MLR_50epoch100mp.tar' % init_lr)
    print("epoch: %d train_loss: %f train_acc: %f validate_loss: %f validate_acc: %f lr: %f" % (epoch+1, losses[-1], acces[-1], validate_loss[-1], validate_acc[-1], scheduler.get_last_lr()[0]))
    f.write("epoch: %d train_loss: %f train_acc: %f validate_loss: %f validate_acc: %f lr: %f\n" % (epoch+1, losses[-1], acces[-1], validate_loss[-1], validate_acc[-1], scheduler.get_last_lr()[0]))
    acc_all[epoch,:] /= validate_fine.shape[0] / num_class
    f1.write("%d\t" % epoch)
    for i in range(num_class):
        f1.write("%f\t" % acc_all[epoch, i])
    f1.write("\n")
    if ((epoch+1) % 10 == 0):
        f.write("Total epoch is %d, the max accuracy is %f, when epoch is %d.\n" % (epoch+1, maxacc, maxepoch+1))
    f.flush()
    f1.flush()
    scheduler.step()
print("Done Training!")
f.write("Done!\n")
f.write("The max accuracy is %f, when epoch is %d.\n" % (maxacc, maxepoch+1))
f.close()
acc_mean = np.mean(acc_all[20:,:],axis=0)
f1.write("mean:\t")
for i in range(num_class):
    f1.write("%f\t"%acc_mean[i])
f1.write("\n")
f1.close()
```

```python
#粗分类，分析训练结果，得到各个类别的准确率并保存图像
if torch.cuda.is_available():
    device = torch.device("cuda:0")
else:
    device = torch.device("cpu")
AUC = dict()
coarse_path = "weights/coarse/"
files = os.listdir(coarse_path)
rnd = 1
for file in files:
    if not os.path.isdir(file):
        if "resnet18" in file:
            net = resnet18(num_classes=20)
        elif "resnext50" in file:
            net = resnext50_32x4d(num_classes=20)
        elif "resnet34" in file:
            net = resnet34(num_classes=20)
        elif "resnext101" in file:
            net = resnext101_32x8d(num_classes=20)
        else:
            continue
    net = net.to(device)
    path = coarse_path + file
    if os.path.exists(path):
        checkpoints = torch.load(path, map_location='cuda:0')
        net.load_state_dict(checkpoints['model_state_dict'])
        loss = checkpoints['loss']
        print("Get Model " + path)
    label_all = None
    pred_all = None
    pred_all_pro = None
    test_loss = 0.0
    test_acc = 0.0
    net.eval()
    with torch.no_grad():
        for data in test_coarse_loader:
            X, label = data
            X = X.float()
            X = Variable(X)
            label = Variable(label)
            X = X.to(device)
            label = label.to(device)
            testout = net(X)
    
            lossess = loss(testout, label)
            test_loss += float(lossess)
            
            _, pred = testout.max(1)
            
            num_correct = (pred == label).sum()
            acc = num_correct
            test_acc += acc
            
            if label_all is None:
                label_all = label
            else:
                label_all = torch.cat([label_all, label])
    
            if pred_all is None:
                pred_all = torch.cat([torch.sigmoid(testout)])
            else:
                pred_all = torch.cat([pred_all, torch.sigmoid(testout)])
    print("loss on test: %f, accuracy on test: %f" % (test_loss / len(test_coarse_loader), test_acc / test_coarse.shape[0]))
    y_test = label_all.cpu().detach().numpy()
    y_pred = pred_all.cpu().detach().numpy()
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    average_precision = dict()
    for i in range(20):
        y_test2 = copy.deepcopy(y_test)
        y_test2[y_test2 != i] = 20
        y_test2[y_test2 == i] = 1
        y_test2[y_test2 == 20] = 0
        y_pred2 = y_pred[:, i]
        fpr[i], tpr[i], _ = roc_curve(y_test2, y_pred2)
        roc_auc[i] = roc_auc_score(y_test2, y_pred2)
    
    AUC[rnd - 1] = roc_auc.copy()
    
    colors = ['b', 'g', 'r', 'k', 'c', 'm', 'y', '#FF7F50', '#524C90', '#845868', '#B0E0E6', '#DA70D6', '#DC143C', '#C0C0C0', '#FFF5EE', '#800000', '#FFD700', '#DEB887', '#6495ED', '#7FFF00']
    plt.figure()
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    for i in range(20):
        plt.plot(fpr[i], tpr[i], label="class" + str(i) + ':ROC curve (area = %0.3f)' % roc_auc[i], color=colors[i])
    plt.legend(loc="lower right")
    plt.savefig(("./plot/coarse/roc_" + file +"_loss_%f_accuracy_%f.jpg") % (test_loss / len(test_coarse_loader), test_acc / test_coarse.shape[0]))
    rnd += 1
```
